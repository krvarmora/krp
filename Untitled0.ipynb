```
# import the libraries 
import tweepy
from textblob import TextBlob 
from wordcloud import WordCloud 
import pandas as pd 
import numpy as np
import re 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
path = "/content/drive/My Drive/data/Cowel.csv"
```

```
df_bonus = pd.read_csv(path)
df_bonus
```

```
df = df_bonus
df.head()
```

```
df["review_text"]
```

```
def cleanTxt (text):
  text = re.sub(r'@[A-Za-z0-9]+ ', '', str(text))
  return text
df['review_text']=df['review_text'].apply(cleanTxt)
```

```
#create a function to get the subjectivity
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity
```

```
#create a function to get the polarity
def getPolarity(text):
  return TextBlob(text).sentiment.polarity
```

```
#create two new columns
df["Subjectivity"] = df["review_text"].apply(getSubjectivity)
df["Polarity"] = df["review_text"].apply(getPolarity)
df
```

```
df1 = df.loc[:,["review_text","Subjectivity","Polarity"]]
df1
```

```
#plot word cloud

allWords = ' '.join([twts for twts in df['review_text']])
wordCloud = WordCloud(width=500, height=300, random_state=21, max_font_size= 119).generate(allWords)
plt.imshow(wordCloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```
#create a function to compute the analysis

def getAnalysis (score):
  if score < 0:
    return 'Negative'
  elif score==0:
    return 'Neutral'
  else:
    return 'Positive'
df1['Analysis'] = df1['Polarity'].apply(getAnalysis)
df1
```

```
#plot the polarity and subjectivity
 
plt.figure (figsize=(8,6))
for i in range(0, df1.shape[0]):
  plt.scatter(df1['Polarity'][i], df1['Subjectivity'][i], color= 'Blue')

plt.title('Coway Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```
